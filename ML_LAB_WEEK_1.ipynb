```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
```

```python
from google.colab import files
data_to_load=files.upload()
```

```python
df=pd.read_csv(io.BytesIO(data_to_load['week1.csv']))
print(df)
```

```python
df.describe()
```

```python
plt.xlabel("area")
plt.ylabel("salesprice")
plt.title("plot scale for salesprice vs area ")
plt.plot(df["area"],df["salesprice"])
```

```python
plt.title("Histogram for saleprice")
plt.xlabel("area")
plt.ylabel("salesprice")
df.plot.hist(bins=10,width=10,color="red")
```

```python
a=np.log(df["area"])
b=np.log(df["salesprice"])
plt.title("logarithm for salesprice")
plt.plot(a,b)
```
